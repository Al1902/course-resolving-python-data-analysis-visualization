## From absolute to the thousands

Don't show all digits of the thousands. For example: 20 000 should be 20.

### Dummy solution

```python
(dff
 .style
    .background_gradient()
    .format(precision=2)
)
```

### Dummy solution

```python
dff = dff.div(1000)

(dff
 .style
    .background_gradient()
    .format(precision=2)
)
```
